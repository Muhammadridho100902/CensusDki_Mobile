```python
import pandas as pd
import numpy as np

# Buat dataset dengan 5 kolom dan 10 baris
dates = pd.date_range('20230101', periods=10)
df = pd.DataFrame(np.random.randn(10, 5), index=dates, columns=['Kolom1', 'Kolom2', 'Kolom3', 'Kolom4', 'Kolom5'])

# Tampilkan dataset
print(df)
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Membuat model ARIMA dengan dataset
model = ARIMA(df['Kolom1'], order=(1,1,1))
model_fit = model.fit()

# Memprediksi nilai Kolom1 untuk 5 data ke depan
pred = model_fit.predict(start=len(df), end=len(df)+4, dynamic=False)

# Tampilkan hasil prediksi
print(pred)
```

```python
import matplotlib.pyplot as plt

# Membuat plot
plt.plot(df.index, df['Kolom1'], label='Data Asli')
plt.plot(pred.index, pred.values, label='Prediksi')

# Menambahkan judul dan label pada sumbu x dan y
plt.title('Prediksi Nilai Kolom1')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Kolom1')
plt.xticks(rotation = 90)

# Menampilkan legenda pada plot
plt.legend()

# Menampilkan plot
plt.show()
```
